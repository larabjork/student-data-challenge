```python
# Import dependencies
import pandas as pd
import numpy
import os
```

```python
# Deliverable 1
# 1.1 Import data from .csv
student_data = os.path.join("Resources", "new_full_student_data.csv")
student_df = pd.read_csv(student_data)
# 1.2 Confirm import with head function
student_df.head()
```

```python
# Deliverable 2
# 2.1 Check for rows with missing values
student_df.isnull().sum()
```

```python
# Drop rows with missing values (in new DataFrame)
no_missing_df = student_df.dropna()
```

```python
# Confirm no missing values
no_missing_df.isnull().sum()
```

```python
# 2.2 Check for duplicated values
no_missing_df.duplicated().sum()
```

```python
# Drop rows with duplicate values (in new DataFrame)
deduped_df = no_missing_df.drop_duplicates()
```

```python
# Check new DataFrame to confirm no duplicated rows
deduped_df.duplicated().sum()
```

```python
# 2.3 Check data types for columns
clean_student_data_df = deduped_df.copy()
clean_student_data_df.dtypes
```

```python
# 2.4 Remove "th" suffix from values in grade column
clean_student_data_df["grade"] = clean_student_data_df["grade"].str.replace("th","")
clean_student_data_df
```

```python
# 2.5 Change data type for grade column
clean_student_data_df["grade"] = clean_student_data_df["grade"].astype(int)

# Check data types for columns
clean_student_data_df.dtypes
```

```python
# Deliverable 3
# 3.1 Generate summary statistics
clean_student_data_df.describe()
```

```python
# 3.2 Display mean math score
clean_student_data_df["math_score"].mean()
```

```python
# 3.3 Store minimum reading score in min_reading_score
min_reading_score = clean_student_data_df["reading_score"].min()
min_reading_score
```

```python
# Deliverable 4
# 4.1 Display grade column by using loc function
clean_student_data_df.loc[:, "grade"]
```

```python
# 4.2 Display 1st 3 rows of columns 3-5, using iloc function
clean_student_data_df.iloc[0:3,3:6]
```

```python
# 4.3 Select rows for 9th graders and display summary stats
ninth_grade_summary = (clean_student_data_df.loc[clean_student_data_df["grade"] == 9]).describe()
ninth_grade_summary
```

```python
# 4.4 Store the row with min reading score in min_reading_row, using loc function and min_reading_score var
min_reading_row = clean_student_data_df.loc[clean_student_data_df["reading_score"] == min_reading_score]
min_reading_row
```

```python
# 4.5 Select all the reading scores from 10th graders at Dixon HS, using loc function with conditionals
filter = clean_student_data_df.loc[(clean_student_data_df["grade"]== 10) & 
                                   (clean_student_data_df["school_name"] == "Dixon High School")]
filter.loc[:,["school_name","reading_score"]]
```

```python
# 4.6 Find mean of reading scores for all 11th and 12th graders, using loc or iloc function
mean_reading_11_12_df = (clean_student_data_df.loc[clean_student_data_df["grade"]>= 11]).loc[:,["reading_score"]].mean()
mean_reading_11_12_df
```

```python
# Deliverable 5
# 5.1 Display the average budget for each school type by using groupby and mean functions
avg_school_budget_by_type_df = (clean_student_data_df.groupby("school_type").mean()).loc[:,["school_budget"]]
avg_school_budget_by_type_df 
```

```python
# 5.2 Find total # of students per school, and sort from largest to smallest
get_all_totals = clean_student_data_df.groupby("school_name").count()
get_all_totals["student_count"]  = get_all_totals.loc[:,["student_id"]]
get_all_totals.loc[:,["student_count"]].sort_values(by = "student_count", ascending = False)
```

```python
# 5.3 Find the average math score by grade for each school type using groupby and mean functions
math_score_by_type_grade_df = (clean_student_data_df.groupby(["school_type", "grade"]).mean().round()).iloc[:,2:3]
math_score_by_type_grade_df
```
